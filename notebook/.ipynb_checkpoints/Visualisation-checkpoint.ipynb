## Visualize events

```python
def zero_crossings(voltage): 
    return np.where(np.diff(np.sign(voltage)))[0]


def get_activations(c, v, Ts=500):
    current, voltage = np.copy(c), np.copy(v)
    zc=select_zc(voltage)
    print(zc)
    k = None
    for j in range(2, len(zc)-2):
        ts=zc[-j]-zc[-(j+2)]
        I=c[zc[-(j+2)]:zc[-j]]
        ic=zero_crossings(I)
        if ts==Ts and len(ic)>1 and len(I)>=Ts:
            k = j
            break

    if k is not None:
        voltage = voltage[zc[-(k+2)]:zc[-k]]
        current = current[zc[-(k+2)]:zc[-k]]
        correct = [zc[-(k+2)], zc[-k]]
    else:
        voltage = voltage[:Ts]
        current = np.zeros((Ts))
        correct = []
    return current, voltage, correct
```

```python
folder = "FINALAGGREGATED"
file1 = folder + '_events'
file2 = folder + '_labels'
data_path='../../data/Data/'
i=1
Ts=500
Nc=5
data = []
labels = []
on_event = []
t, f = select_data_appliance_type(folder, i, data_path, Ts, Nc)
data += t
labels += [i] * len(t)
if i != 9:
    on_event+= [1, 0] * int(len(t) / 2)
else:
    on_event+= [1, 1, 0] * int(len(t) / 3)
```

```python
plt.plot(t[0][Ts*Nc:,1])
```

```python
idx=-1
c_b, v_b, zc_b=get_activations(t[idx][:Ts*Nc,0],t[idx][:Ts*Nc,1], Ts)
c_a, v_a, zc_a=get_activations(t[idx][Ts*Nc:,0],t[idx][Ts*Nc:,1], Ts)
event = on_event[idx]
if c_a.any() and event==1:
        e_c = c_a - c_b
        e_v = v_a
elif c_b.any() and event==0:
        e_c = c_b - c_a
        e_v = v_b
plt.title(event)
plt.plot(c_a, label="ca")
plt.plot(c_b, label="cb")
plt.plot(e_c, label="ce")
plt.legend()
```
